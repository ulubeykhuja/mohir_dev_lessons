```python
(df["price"] == ('Договорная')).sum()
```

```python
# Договорная' qiymatiga ega qatorlarni o'chirish
df = df[df["price"] != ('Договорная')]
```

```python
df = df[df["price"] >2000]
```

```python
# 'price' ustunini raqamli turga o'tkazish, xatolarni NaN ga o'tkazish
df['price'] = pd.to_numeric(df['price'], errors='coerce')
```

```python
# NaN qiymatlarga ega qatorlarni o'chirish (konvertatsiya qilib bo'lmagan qiymatlar)
df.dropna(subset=['price'], inplace=True)
```

```python
# Убедимся, что столбец 'price' теперь является числовым
if not pd.api.types.is_numeric_dtype(df['price']):
    print("Ошибка: Столбец 'price' не является числовым после преобразования и удаления NaN.")
    # Вы можете захотеть проверить оставшиеся нечисловые значения, если появится это сообщение
    # print(df[~pd.to_numeric(df['price'], errors='coerce').notna()])
else:
    print("Столбец 'price' успешно преобразован в числовой.")
```
