## SQL buyruqlarini f-string yordamida yozish


```python
df = pd.read_csv("world_population.csv")
df.head()
```

```python
query = """ 
CREATE TABLE IF NOT EXISTS worldpop (
rank INT,
country TEXT PRIMARY KEY,
density DOUBLE,
population DOUBLE NOT NULL,
area INT NOT NULL
)
"""
cursor.execute(query)
db.commit()
```

```python
uz = df[df.country == 'Uzbekistan']
uz
```

```python
rank = uz.index.values[0]
countru = uz.at[rank, 'country']
density = uz.at[rank, 'density']
population = uz.at[rank, 'pop2021']
area = uz.at[rank, 'area']
print(f"{rank}, {countru}, {density}, {population}, {area}")
```

```python
query = f"""
INSERT INTO worldpop VALUES ({rank}, '{countru}', {density}, {population}, {area})
"""
cursor.execute(query)
db.commit()
```

```python
pd.read_sql("SELECT * FROM worldpop", db)
```

## Between, in va like

```python
query = "SELECT * FROM elect WHERE country = 'Uzbekistan' AND year BETWEEN 2005 AND 2015"
pd.read_sql(query, db)
```

```python
query = "SELECT * FROM elect WHERE country = 'Uzbekistan' AND year IN (1990,1995,2000,2005)"
pd.read_sql(query, db)
```

```python
# LIKE
# % - har qanday belgilar
# _ bilan bo'lsa bittq harf mos kelsa qaytaradi 
query = "SELECT * FROM elect WHERE country LIKE '%stan' AND year IN (2019,2020,2021)"
pd.read_sql(query, db)
```

```python
query = "SELECT * FROM elect WHERE code LIKE 'A__' AND year IN (1990,1995,2000,2005)"
pd.read_sql(query, db)
```
