```python
x = np.arange(0, 4*np.pi, 0.01)
y = np.sin(x)
```

```python
plt.plot(x, y, )
plt.show()
```

```python
x = np.arange(0, 4*np.pi, 0.01)
sin = np.sin(x)
cos = np.cos(x)

plt.figure(figsize=(12, 5))

plt.plot(x, sin,
         linewidth=1.5,
         linestyle='--',
         color='red',
         alpha=0.7,
         marker = 'h')
plt.plot(x, cos,
         linewidth=1.5,
         linestyle=':',
         color='blue',
         alpha=0.7,
         marker = 'v')

plt.title('sin(x) va cos(x)')
plt.xlabel('pi')
plt.legend(['sin(x)', 'cos(x)'], loc=0, fontsize='x-large')
plt.grid(color = 'tomato', alpha=0.5, linestyle='--')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/museum_visitors.csv', index_col=0)
df.head()
```

```python
sns.lineplot(data=df)
```

```python
plt.figure(figsize=(12, 5))
plt.title("Muzegga tashriflar")
plt.xlabel("Sana")
plt.xticks(rotation=90)
plt.ylabel("Odamlar soni")
plt.grid()
sns.lineplot(data=df)
plt.show()
```
