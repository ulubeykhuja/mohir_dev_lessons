```python
x = np.arange(0, 4*np.pi, 0.01)
sin = np.sin(x)
cos = np.cos(x)

plt.figure(figsize=(12, 5))

plt.plot(x, sin,
         linewidth=1.5,
         linestyle='--',
         color='red',
         alpha=0.7,
         marker = 'h')
plt.plot(x, cos,
         linewidth=1.5,
         linestyle=':',
         color='blue',
         alpha=0.7,
         marker = 'v')

plt.title('sin(x) va cos(x)')
plt.xlabel('pi')
plt.legend(['sin(x)', 'cos(x)'], loc=0, fontsize='x-large')
plt.grid(color = 'tomato', alpha=0.5, linestyle='--')
plt.show()
```

HEATMAP

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/museum_visitors.csv', index_col=0, parse_dates=[0], usecols=[0,1])
df.head()
```

```python
df['month'] = [i.month for i in df.index]
df.head()
```

```python
df['year'] = [i.year for i in df.index]
df.head()
```

```python
df2 = df.groupby(['month', 'year']).max()
df2.head()
```

```python
df2.unstack(level=0)
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df2.unstack(level=0), cmap='YlGnBu', linewidths=1, square=True)
xticks_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(np.arange(12) + .5, labels=xticks_labels)
plt.show()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/merc.csv')
df.shape
```

```python
df = df[df.year>=2010]
df = df[df.transmission!='Other']
df = df[df.fuelType!='Other']
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='price')
plt.grid()
plt.show()
```

```python
sns.displot(data=df, x='price', hue='transmission', height=5, aspect=2)
```

```python
sns.displot(data=df, x='price',col='transmission', height=3, aspect=2)
```
