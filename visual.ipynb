```python
x = np.arange(0, 4*np.pi, 0.01)
sin = np.sin(x)
cos = np.cos(x)

plt.figure(figsize=(12, 5))

plt.plot(x, sin,
         linewidth=1.5,
         linestyle='--',
         color='red',
         alpha=0.7,
         marker = 'h')
plt.plot(x, cos,
         linewidth=1.5,
         linestyle=':',
         color='blue',
         alpha=0.7,
         marker = 'v')

plt.title('sin(x) va cos(x)')
plt.xlabel('pi')
plt.legend(['sin(x)', 'cos(x)'], loc=0, fontsize='x-large')
plt.grid(color = 'tomato', alpha=0.5, linestyle='--')
plt.show()
```
