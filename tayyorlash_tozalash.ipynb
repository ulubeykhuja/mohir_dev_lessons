## .fillna() – NaN qiymatlarni toʻldirish

```python
df.fillna(0)
```

```python
df.describe()
```

```python
df.fillna(0).describe()
```

```python
df.fillna({'horsepower': 100, 'wheel-base': 90})
```

```python
# O'rtacha qiymatlarni qo'yish
df.fillna({'horsepower': df.horsepower.mean(), 'wheel-base': df['wheel-base'].mean()})
```

```python
# avvalgi qiymatlar bilan to'dirish
df.fillna(method='ffill')
```

```python
# keyingi qatorni qo'yish
df.fillna(method='bfill')
```

```python
df[df.company == 'toyota'].bfill()
```

```python
# barcha qiymatlarni mean bilan almashtirish
df.fillna(df.mean(numeric_only=True))
```
