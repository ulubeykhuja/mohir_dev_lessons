## .cut() va .qcut() yordamida guruhlash

```python
df = pd.read_csv(
    r"c:\Users\User\projects\MohirDev\praktikum_datasets-main\world_population.csv", 
    usecols=['country', 'pop2021', 'area'],
    index_col='country')
df.head()
```

```python
df.loc['Uzbekistan']
```

```python
df.loc['India']
```

```python
bins = [0, 1000, 10_000, 30_000, 50_000, 100_000, 300_000, 500_000, 1_500_000]
```

```python
guruh = pd.cut(df.pop2021, bins)
guruh
```

```python
guruh.value_counts()
```

```python
df['guruh'] = guruh
df
```

```python
df = pd.read_csv(
    r"c:\Users\user\1. MohirDev\praktikum_datasets-main\automobile_data.csv", 
    index_col=0)
df.head()
```

```python
bins = [0, 10000, 20000, 50000]
```

```python
oraliq = pd.cut(df.price, bins)
oraliq
```

```python
oraliq.value_counts()
```

```python
names = ["arzon", "o'rta", "qimmat"]
oraliq = pd.cut(df.price, bins, labels=names)
oraliq.value_counts()
```

```python
range = pd.cut(df.price, 3)
range.value_counts()
```

```python
# .qcut() - taxminan bir xil miqdorda bo'ladi
range = pd.qcut(df.price, 3)
range.value_counts()
```

```python
range = pd.qcut(df.horsepower, 3)
range.value_counts()
```
